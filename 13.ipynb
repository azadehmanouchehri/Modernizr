```
import zipfile
zip_ref = zipfile.ZipFile('my_data/gray im test.zip', 'r')
zip_ref.extractall('data1')
zip_ref.close()
import cv2
import numpy as np
import glob
image_path="data1/gray im test/"
images=glob.glob(image_path+"*.jpg")
images.sort()
test_data=[]
for img in images:
    image=cv2.imread(img)
    image=cv2.resize(image,(256, 256))
    #image=image.astype(np.float32)
    test_data.append(image)
test_data = np.reshape(test_data, (len(test_data), 256, 256, 3))
```

```
import cv2
import numpy as np
import glob
import zipfile
zip_ref = zipfile.ZipFile('my_data/test GT.zip', 'r')
zip_ref.extractall('GT1')
zip_ref.close()
lebel_path="GT1/test GT/"
lebels=glob.glob(lebel_path+"*.jpg")
lebels.sort()
y_test=[]
i=0
seg_labels = np.zeros([len(lebels),256, 256,2], dtype='uint8')
for l in lebels:
    label=cv2.imread(l)
    label=cv2.resize(label,(256, 256))
    label=label.astype(np.float32)
    label = label[:,:,0]
    seg_labels[i,:,:,1]=(label>=26)
    seg_labels[i,:,:,0]=(label<=25)
    y_test.append(seg_labels[i])
    i=i+1

y_test = np.reshape(y_test, (len(y_test), 256, 256,2))

```

```
import matplotlib.pyplot as plt
from skimage import measure
TP=0
FP=0
FN=0
for i in range(196):
  
  L=y_test[i,:,:,0]
  l=measure.label(L, background=True, connectivity=2)
  a1=np.unique(l)
  b1=np.max(a1)
  
  predict_im=decoded_imgs[i,:,:,0]
  #predict_im1=decoded_imgs[i,:,:,1]
  #predict_im=np.zeros([256, 256], dtype='uint8')
  #for j in range(255):
   # for k in range(255):
   #   if (decoded_imgs[i,j,k,1]>=decoded_imgs[i,j,k,0]):
    #    predict_im[j,k]=1
        
  
  m=np.min(predict_im)
  predict_im=(predict_im>m + 0.34)
  I=measure.label(predict_im, background=True, connectivity=2)
  a2=np.unique(I)
  b2=np.max(a2)
  for j in range(b2):
    t=j+1
    I1=(I==t)
    s1=I1.sum()
    if(s1<200):
      predict_im[I==t]=1
  
  I=measure.label(predict_im, background=True, connectivity=2)
  a2=np.unique(I)
  b2=np.max(a2)   
    
  
  
  
  C=np.zeros([b1, b2], dtype='uint8')
  for j in range(b1):
    t=j+1
    G=(l==t)
    sG=G.sum()
    for k in range(b2):
        q=k+1
        P=(I==q)
        SP=P.sum()
        m=min(sG,SP)
        T=np.multiply(P, G)
        S=T.sum()
        if(S>(m/10)):
          C[j,k]=1
          
  for j in range(b1):
    a=C[j,:].sum()
    if(a<1):
      FN=FN+1
    else:
      TP=TP+1
  for k in range(b2):
    a=C[:,k].sum()
    if(a<1):
      FP=FP+1
        
print('TP:',TP)
print('FP:',FP)
print('FN:',FN)
Prec=TP/(TP+FP)
print('Prec:',Prec)
Rec=TP/(TP+FN)
print('Rec:',Rec)
#Spec=TN/(FP+FN)
F1=(2*Prec*Rec)/(Prec+Rec)
print('F1:',F1)
F2=(5*Prec*Rec)/(4*Prec+Rec)
print('F2:',F2)
        
        
        
        
        
      
 
```
